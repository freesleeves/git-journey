## NLP Pipeline using Spacy

```python
import spacy
```

```python
text='In Vision Studio, you can find your resource details such as its name and pricing tier as well as switch resources by selecting the Settings icon in the top right corner of the Vision Studio screen.'
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(text)
```

```python
for token in doc:
  print(token.text)
```

```python
#remove stop words
tokens=[token.text for token in doc if not token.is_stop]
print(tokens)
```

```python
# remove punctuations and stopwords
filtered_tokens=[token.text for token in doc if not token.is_punct and not token.is_stop]
print(filtered_tokens)
```

```python
for token in doc:
  if not token.is_stop and not token.is_punct:
    print(token.text,token.lemma_,token.pos_, token.tag_, token.dep_,[(ent.text, ent.label_) for ent in doc.ents])
```

```python
for ent in doc.ents:
  print(ent.text,ent.label_)
```
